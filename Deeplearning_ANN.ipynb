```python
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('/content/titanic_cleaned - titanic_cleaned.csv')
```

```python
df.head()
```

```python
y = df['Survived']
X = df.drop('Survived', axis=1)
```

```python
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.2,train_size=0.8,random_state=42)
```

SCALING

```python
s= StandardScaler()
X_train= s.fit_transform(X_train)
X_test=s.transform(X_test)
```

CREATE NEURONS

 y = mx+c --- c - bias,
 if there is error

```python
model= tf.keras.models.Sequential([tf.keras.Input(shape=(X_train.shape[1],)),
                                   tf.keras.layers.Dense(16,activation='relu'), #Dense-hidden layer
                                   tf.keras.layers.Dense(1,activation='sigmoid')])
```

```python
X_train.shape
```

```python
X_train.shape[0]
```

```python
X_train.shape[1]
```

```python
#optimizer -get an parameters to updates
#
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

epochs- while backpropagation sample... There is a multiple samples ..

```python
from keras import callbacks
early_stopping= callbacks.EarlyStopping(min_delta = 0.0001 ,patience= 70)
```

if user should not give early stopping then it should be overfitting

```python
model.fit(X_train,y_train,epochs= 60,validation_split=0.2,callbacks =early_stopping)
```

```python
loss, accuracy = model.evaluate(X_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
y_pred= model.predict(X_test)
```

```python
y_pred =(y_pred>0.5).astype(int)
print(classification_report(y_pred,y_test))
```

```python
confusion_matrix(y_pred,y_test)
```
