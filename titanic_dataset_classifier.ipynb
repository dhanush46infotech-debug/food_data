```python
import pandas as pd
import matplotlib.pyplot as plt #visualization

```

```python
file_location = "/content/titanic_cleaned.xlsx"
df_titanic = pd.read_excel(file_location)
```

```python
df_titanic.head()
```

```python
x = df_titanic.drop('Survived',axis=1)
y = df_titanic['Survived']
```

```python
from sklearn.model_selection import train_test_split


## train test split data
X_train,X_test,y_train,y_test=  train_test_split(x,y,test_size=0.2,train_size=0.8,random_state=100)
```

```python
## scale the model
from sklearn.preprocessing import StandardScaler
s= StandardScaler()
X_train=s.fit_transform(X_train)
X_test=s.transform(X_test)
```

RANDOM FOREST CLASSIFIER (WITH DETAIL NOTES)

```python
#RANDOM FOREST CLASSIFIER
from sklearn.ensemble import RandomForestClassifier
modeldf = RandomForestClassifier()
modeldf.fit(X_train, y_train)
```

```python
preddf =modeldf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
print(accuracy_score(y_test,preddf))

#TRUE_POSITIVE + TRUE_NEGATIVE/ TOTAL_ROWS
```

```python
  #SINCE MY DATASET HAS (890 ROWS)
  #TRAINING - 80%
  #TESTING - 20%
  #WHICH MEANS -- 890/20% = 180
  #[[88 16]
 #[20 55]]  ----- 180 (IF USER ADDS THESE 4 --> GET THIS TOTAL OF THAT VALUE)
#TOTAL ROWS
print(confusion_matrix(y_test,preddf))
```

```python
print(classification_report(y_test,preddf))
#PRECISION -POSITIVE PREDICTION
#RECALL - NEGATIVE PREDICTION
#F1-SCORE - BALANCE BETWEEN PRECISION & RECALL
```

DECISION TREE CLASSIFIER

```python

from sklearn.tree import DecisionTreeClassifier

modeldf2= DecisionTreeClassifier()
modeldf2.fit(X_train,y_train)
```

```python
## predict
y_preddf2=modeldf2.predict(X_test)
```

```python
## Accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_preddf2)
```

```python
#confusion_matrix
confusion_matrix(y_test,y_preddf2)
```

```python
#classification
from sklearn.metrics import classification_report
print(classification_report(y_test,y_preddf2))
```

```python
from sklearn.neighbors import KNeighborsClassifier
modeldf3= KNeighborsClassifier()
modeldf3.fit(X_train,y_train)
```

```python
#to Predict
x_preddf3 =modeldf3.predict(X_test)
```

```python
accuracy_score(y_test,x_preddf3)
```

```python
confusion_matrix(y_test,x_preddf3)
```
